```python
import cv2
import datetime
import math
import numpy as np
```

### Analog Clock

```python
colors = {'blue': (255, 0, 0), 'green': (0, 255, 0), 'red': (0, 0, 255), 'yellow':
(0, 255, 255), 'magenta': (255, 0, 255), 'cyan': (255, 255, 0), 'white': (255, 255,
255), 'black': (0, 0, 0), 'gray': (125, 125, 125), 'rand': np.random.randint(0,
high=256, size=(3,)).tolist(), 'dark_gray': (50, 50, 50), 'light_gray': (220, 220,
220)}
```

```python
img = np.zeros((700,700,3), dtype = 'uint8')
img[:] = colors['gray']
cv2.circle(img, (350,350), 5, color = colors['black'], thickness = -1)
cv2.circle(img, (350,350), 330, color = colors['black'], thickness = 2)

cv2.putText(img, 'Welcome To CV', 
            (200,100), cv2.FONT_HERSHEY_COMPLEX, 0.9, colors['black'],2, 4)
constant = 350
r = 330
for i in range(12):
    x = constant + r* math.cos((-270-(30*i))*(np.pi/180))
    y = constant + r* math.sin((-270-(30*i))*(np.pi/180))
    cv2.circle(img, (int(x),int(y)), 4, color = colors['black'], thickness = -1)

image = img.copy()
while True:
    sec = datetime.datetime.now().second
    mini = datetime.datetime.now().minute
    hour = datetime.datetime.now().hour
    if hour >12:
        hour = hour-12
    x_sec = round(constant + (r-40)* math.cos((-90+(6*sec))*(np.pi/180)))
    y_sec = round(constant + (r-40)* math.sin((-90+(6*sec))*(np.pi/180)))
    
    x_min = round(constant + (r-60)* math.cos((-90+(6*mini))*(np.pi/180)))
    y_min = round(constant + (r-60)* math.sin((-90+(6*mini))*(np.pi/180)))
  
    x_hour = round(constant + (r-100)* math.cos((-90+(30*hour))*(np.pi/180)))
    y_hour = round(constant + (r-100)* math.sin((-90+(30*hour))*(np.pi/180)))
    
#     print(datetime.datetime.now().second -1)
    
    cv2.line(img, (350,350),(x_sec, y_sec), color = colors['red'],thickness = 2)
    cv2.line(img, (350,350),(x_min, y_min), color = colors['red'],thickness = 4)

    cv2.line(img, (350,350),(x_hour, y_hour), color = colors['red'],thickness = 8)

    cv2.imshow('Image', img)
    img = image.copy()
    if cv2.waitKey(500) & 0xFF == ord('q'):
        break
cv2.destroyAllWindows() 
```
