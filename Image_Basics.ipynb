## Reading webcamera

```python
capture = cv2.VideoCapture(0)
```

```python
# Need to check is connection established or not, Perfect we are connected to webcam
while capture.isOpened():
    # Capture frame by frame and then we will show.
    ret, frame = capture.read()
    if ret:
        cv2.imshow('Image from webcam', frame)
        
        grey_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
        cv2.imshow('Grey Image', grey_frame)
        
        if cv2.waitKey(20) & 0xFF == ord('q'):
            break
    else:
        break
capture.release()
cv2.destroyAllWindows()
```

```python
import numpy as np
import matplotlib.pyplot as plt
colors = {'blue': (255, 0, 0), 'green': (0, 255, 0), 'red': (0, 0, 255), 'yellow':
(0, 255, 255), 'magenta': (255, 0, 255), 'cyan': (255, 255, 0), 'white': (255, 255,
255), 'black': (0, 0, 0), 'gray': (125, 125, 125), 'rand': np.random.randint(0,
high=256, size=(3,)).tolist(), 'dark_gray': (50, 50, 50), 'light_gray': (220, 220,
220)}
```

```python
def show_with_matplotlib(img, image_name):
    plt.figure(figsize = (6,12),)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.title(image_name)
    plt.imshow(img)
```

```python
image  = np.zeros((400,400, 3), dtype = 'uint8')
image[:] = colors['gray']
```

```python
seperation = 40
show_with_matplotlib(cv2.line(image, (0,seperation), (400, seperation),10, lineType = cv2.LINE_8), 'Image')
```

### Draw line

```python
image  = np.zeros((400,400, 3), dtype = 'uint8')
image[:] = colors['gray']
cv2.line(image, (0,0), (400,400), color =colors['green'], thickness=5,lineType = 8)
cv2.line(image, (350,0), (400,400), color =colors['green'], thickness=5,lineType = 8)

show_with_matplotlib(image, 'cv2_Line')
```

### Draw rectangle

```python
img = np.zeros((500,500,3), dtype = 'uint8')
img[:] = colors['gray']
```

```python
cv2.rectangle(img, (10,10), (30,400), color = colors['blue'], thickness=-1)

cv2.rectangle(img, (100,10), (300,400), color = colors['red'], thickness=10)

show_with_matplotlib(img, 'Recttangle')
```

### Draw Circle

```python
img = np.zeros((500,500,3), dtype = 'uint8')
img[:] = colors['gray']
```

```python
cv2.circle(img, (250,250),250, color = colors['blue'])

show_with_matplotlib(img, 'Circle')
```

### Draw Arrowline

```python
img = np.zeros((500,500,3), dtype = 'uint8')
img[:] = colors['gray']
```

```python
cv2.arrowedLine(img, (10,10), (100,400),color = colors['red'], thickness = 5, line_type = 8)
show_with_matplotlib(img, 'Arrow')
```

### Writing Text

```python
img = np.zeros((500,500,3), dtype = 'uint8')
img[:] = colors['gray']
```

```python
img.fill(255)
cv2.putText(img, 'Learning OpenCv', (10,30), cv2.FONT_HERSHEY_COMPLEX, 0.9, colors['red'],2, 4)

show_with_matplotlib(img, 'Put Text')
```

### Dynamic Drawing Using Mouse

```python
img = np.zeros((500,500,3), dtype = 'uint8')
# img[:] = colors['gray']
# img.fill(255)
# cv2.imshow('Image mouse', img)
cv2.namedWindow('Image mouse')
def draw_circle(event, x,y,flags, param):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        cv2.circle(img, (x, y), 10, colors['red'], -1)
        
    if event == cv2.EVENT_MOUSEMOVE:
        cv2.circle(img, (x, y), 1, colors['red'], -1)
        print("event: EVENT_MOUSEMOVE")
    if event == cv2.EVENT_LBUTTONUP:
        print("event: EVENT_LBUTTONUP")
    if event == cv2.EVENT_LBUTTONDOWN:
        print("event: EVENT_LBUTTONDOWN")

cv2.setMouseCallback('Image mouse', draw_circle)

while True:
    cv2.imshow('Image mouse', img)
    
    if cv2.waitKey(20) & 0xFF== ord('q'):
        break
cv2.destroyAllWindows()
```
